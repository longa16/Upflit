```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('Kevin_Hillstrom_MineThatData_E-MailAnalytics_DataMiningChallenge_2008.03.20.csv')
print(df.head())
```

```python
#identification des clients qui ont recu des 'Mens E-mail'
df_ab = df[df['segment'] != 'Womens E-Mail'].copy()
df_ab['treatment'] = np.where(df_ab['segment'] == 'Mens E-mail', 1, 0)

print(f"Taille de l'échantillon : {len(df_ab)} clients")
print("Répartition des groupes :")
print(df_ab['segment'].value_counts())
```

```python
#evaluation taux de conversion des 'Mens E-mail' 
ab_summary = df_ab.groupby('segment')[['conversion', 'spend']].mean()
ab_summary['uplift_conversion'] = ab_summary['conversion'] - ab_summary.loc['No E-Mail', 'conversion']
ab_summary['upflit_spend'] = ab_summary['spend'] - ab_summary.loc['No E-Mail', 'spend']

print("\n--- RÉSULTATS GLOBAUX ---")
print(ab_summary)
```

```python
# Visualisation
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x=ab_summary.index, y=ab_summary['conversion'], ax=axes[0], palette='pastel')
axes[0].set_title('Taux de Conversion (%)')
axes[0].set_ylabel('Probabilité d\'achat')

sns.barplot(x=ab_summary.index, y=ab_summary['spend'], ax=axes[1], palette='pastel')
axes[1].set_title('Dépense Moyenne ($)')
axes[1].set_ylabel('Dollars dépensés')

plt.show()
```

```python
#two model approach
from sklearn.model_selection import train_test_split


features = ['recency', 'history_segment', 'history', 'mens', 'womens', 'zip_code', 'channel']


df_model = pd.get_dummies(df_ab[features])


X = df_model
y = df_ab['conversion'] # Cible : Achat (0 ou 1)
treat = df_ab['treatment'] # Traitement : A reçu l'email (0 ou 1)


X_train, X_test, y_train, y_test, treat_train, treat_test = train_test_split(
    X, y, treat, test_size=0.2, random_state=42
)

print(f"Features prêtes : {X_train.shape[1]} colonnes")
print(X_train.head())
```

```python
print("--- Noms exacts des segments dans ton fichier ---")
print(df['segment'].unique())

df_ab = df.copy() 

df_ab = df_ab[df_ab['segment'].str.contains("Mens|No E-Mail", case=False, na=False)]

df_ab['treatment'] = np.where(df_ab['segment'].str.contains("Mens", case=False), 1, 0)

print("\n--- Vérification après correction ---")
print(df_ab['treatment'].value_counts())
```

```python
features = ['recency', 'history_segment', 'history', 'mens', 'womens', 'zip_code', 'channel']
df_model = pd.get_dummies(df_ab[features])

X = df_model
y = df_ab['conversion']
treat = df_ab['treatment']

# Split
X_train, X_test, y_train, y_test, treat_train, treat_test = train_test_split(
    X, y, treat, test_size=0.2, random_state=42
)

print(f"Nombre de personnes avec Email dans le Train : {sum(treat_train == 1)}")
```

```python
from sklearn.ensemble import RandomForestClassifier

#Séparation des données d'entraînement
#Ceux qui ont reçu l'email
X_train_T = X_train[treat_train == 1]
y_train_T = y_train[treat_train == 1]

#Ceux qui n'ont rien reçu
X_train_C = X_train[treat_train == 0]
y_train_C = y_train[treat_train == 0]


model_t = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model_t.fit(X_train_T, y_train_T)


model_c = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model_c.fit(X_train_C, y_train_C)

print("Les deux modèles sont entraînés !")

# Calcul de l'Uplift sur le jeu de Test
prob_if_treated = model_t.predict_proba(X_test)[:, 1] # Probabilité d'achat SI email
prob_if_control = model_c.predict_proba(X_test)[:, 1] # Probabilité d'achat SI rien

uplift_score = prob_if_treated - prob_if_control

results = pd.DataFrame({
    'prob_treated': prob_if_treated,
    'prob_control': prob_if_control,
    'uplift_score': uplift_score,
    'true_conversion': y_test,
    'was_treated': treat_test
})

print("\n--- Exemple de résultats pour 5 clients ---")
print(results.head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(results['uplift_score'], kde=True, color='purple')
plt.title('Distribution du Score d\'Uplift')
plt.xlabel('Impact estimé de l\'email (Différence de probabilité)')
plt.ylabel('Nombre de clients')
plt.axvline(0, color='red', linestyle='--', label='Neutre (0)')
plt.legend()
plt.show()

# Compter combien de personnes on devrait cibler 
potential_targets = results[results['uplift_score'] > 0]
print(f"Sur {len(results)} clients testés :")
print(f"- {len(potential_targets)} ont un Uplift Positif (À cibler )")
print(f"- {len(results) - len(potential_targets)} ont un Uplift Négatif ou Nul (À éviter )")
```
